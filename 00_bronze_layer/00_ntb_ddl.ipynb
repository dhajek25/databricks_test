### Define functions

```python
def run_sql_file(file_path: str):
    """
    Reads a .sql file and executes its content in Databricks.
    """
    try:
        with open(file_path, 'r') as f:
            sql_queries = f.read()
            
        # Splitting by ';' in case your script contains multiple statements
        for query in sql_queries.split(';'):
            clean_query = query.strip()
            if clean_query:
                print(f"Executing query from {file_path}...")
                spark.sql(clean_query)
        
        print(f"Successfully executed {file_path}")
    except Exception as e:
        print(f"Error running {file_path}: {e}")
```

### Run DDL scripts

```python
# Define Item table

run_sql_file(ddl_path_item)
```

```python
# Define Event table

run_sql_file(ddl_path_event)
```
